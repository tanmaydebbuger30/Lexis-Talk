```python
import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
sample_texts = {
    "../data/text_files/python_intro.txt":"""Python Programming intro

    Artificial Intelligence (AI) refers to the simulation of human intelligence in machines that are designed to think, learn, and make decisions like humans. AI systems can perform tasks such as speech recognition, image understanding, natural language processing, and decision-making with minimal human intervention.
""",
    "../data/text_files/ml_text.txt":""" ML basics

    Machine Learning (ML) is a subset of AI that enables systems to automatically learn and improve from experience without being explicitly programmed. ML algorithms analyze large volumes of data, identify patterns, and make predictions or classifications. Common types of ML include supervised learning, unsupervised learning, and reinforcement learning.

    Together, AI and ML are transforming industries by enabling automation, predictive analytics, personalized experiences, and data-driven decision-making. They are widely used in fields such as healthcare (disease prediction), finance (fraud detection), marketing (customer segmentation), and technology (recommendation systems and chatbots). """


}
for filepath,content in sample_texts.items():
    with open(filepath,"w") as f:
        f.write(content)
print("sample textt file creeated")
```

```python
###Textloader

from langchain_community.document_loaders import TextLoader, TextLoader, JSONLoader

loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")
document = loader.load()
print(document)
```

```python
### Directory loader
from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",
    loader_cls=TextLoader,
    loader_kwargs={"encoding":"utf-8"},
    show_progress=False,
)

documents = dir_loader.load()
print(documents)
```

```python
### Directory loader
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

dir_loader = DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf",
    loader_cls = PyMuPDFLoader,
    # loader_kwargs={"encoding":"utf-8"},
    show_progress=False,
)

pdf_documents = dir_loader.load()
print(pdf_documents)
```

```python
from langchain_community.document_loaders.csv_loader import CSVLoader


csv_loader = DirectoryLoader(
    "../data/csvs",
    glob="**/*.csv",
    loader_cls=CSVLoader,
    show_progress=False,
)

document_3 = csv_loader.load()
print(document_3)
```

```python
from langchain_community.document_loaders.json_loader import JSONLoader

json_loader = JSONLoader(
    "../data/json/sample.json",
    # glob="**/*.json",
    jq_schema=".[] | tostring",
    text_content=False,
)

document_json = json_loader.load()
print(document_json)
```
